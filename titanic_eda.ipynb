```python
# 📦 Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 📥 Load the Titanic dataset
df = pd.read_csv("titanic.csv")

# 👀 Show the first 5 rows
df.head()
```

```python
# 🧠 Basic Information
print("Dataset Shape:", df.shape)

# Info about data types and nulls
df.info()

# Summary statistics
df.describe(include='all')
```

```python
# 🔍 Check for missing values
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())

```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])  

```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
```

```python
df.isnull().sum()
df.info()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.savefig("images/survival_count.png")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival by Gender")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.legend(title="Sex")
plt.savefig("images/survival_by_gender.png")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival by Gender")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.legend(title="Sex")
plt.savefig("images/survival_by_gender.png")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Survival by Passenger Class")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.legend(title="Pclass")
plt.savefig("images/survival_by_class.png")
plt.show()
```

```python
sns.histplot(data=df, x='Age', kde=True, bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.savefig("images/age_distribution.png")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survival")
plt.xlabel("Survived")
plt.ylabel("Age")
plt.savefig("images/age_vs_survival_boxplot.png")
plt.show()
```

```python
sns.histplot(df['Fare'], kde=True)
plt.title('Fare Distribution')
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")
plt.savefig("images/fare_distribution.png")
plt.show()
```

```python
df.to_csv("titanic_cleaned.csv", index=False)
```
